### What is diag_1, diag_2, diag_3

- There are three column which contains the diagnosis code for the Encounter.
- Each time the patient admits to the hospital, A diagnosis code is assign with it.
- Which means on which problem the patient comes to the hospital.

- Diagnosis 1 :Nominal The primary diagnosis (coded as first three digits of ICD9); 848 distinct values 0%
- Diagnosis 2 :Nominal Secondary diagnosis (coded as first three digits of ICD9); 923 distinct values 0%
- Diagnosis 3 :Nominal Additional secondary diagnosis (coded as first three digits of ICD9); 954 distinct values

As per the Documentation, they are the ICD - 9 Coded Diagnosis Codes, Each code represents some Disease

- The diag_1 is the Primary Diagnosis of the Patient, which means the patient is admitted to the hospital on this diagnosis.
- The diag_2 is the Secondary Diagnosis, According to CMS Documentation "Secondary diagnoses are “conditions that coexist at the time of admission, that develop subsequently, or that affect the treatment received and/or length of stay. These diagnoses are vital to documentation and have the potential to impact a patient's severity of illness and risk of mortality"
- The diag_3 is the Additional Secondary Diagnosis

### Converting Categorical Data into Numerical Representation

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
```

```python
categorical_features =['race', 'gender', 'age', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id', 'diag_1', 'diag_2', 'diag_3', 'number_diagnoses',
       'max_glu_serum', 'A1Cresult', 'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'glipizide', 'glyburide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'insulin', 'glyburide-metformin', 'change', 'diabetesMed'] 

for i in categorical_features:
    df_[i] = le.fit_transform(df_[i])
```

### Splitting the Dependent and Independent Variables

```python
X = df_.drop(columns= ['readmitted_2_encoded'])
Y = df_['readmitted_2_encoded']
```

### Feature Scaling

```python
from sklearn import preprocessing
scaled_X = preprocessing.StandardScaler().fit_transform(X)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_X, Y, test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Modeling

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_confusion_matrix, plot_precision_recall_curve
```

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
# Define Model
lr = LogisticRegression()
# Training
lr.fit(X_train, y_train)

# Prediction
lr_prediction = lr.predict(X_test)


from sklearn.metrics import accuracy_score, mean_squared_error

# Calculate accuracy
accuracy = accuracy_score(y_test, lr_prediction)

# Calculate mean squared error (MSE)
mse = mean_squared_error(y_test, lr_prediction)

# Calculate root mean squared error (RMSE)
rmse = np.sqrt(mse)

# Print the results
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("MSE: {:.4f}".format(mse))
print("RMSE: {:.4f}".format(rmse))

```

```python
print(classification_report(y_test, lr_prediction, target_names= ['Not Readmitted', 'Readmitted']))
```

### Confusion Matrix

```python
fig, ax = plt.subplots(figsize=(12, 4))
plot_confusion_matrix(lr, X_test, y_test,
                      display_labels = ['Not Readmitted', 'Readmitted'],
                      cmap = 'Blues',
                      ax=ax
                    )
plt.title('Confusion Matrix of Logistic Regression Model \n', size = 15)
plt.show()
```

### Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

# Define Model
rf = RandomForestClassifier(n_estimators = 450, max_depth=9, random_state=43)

# Training
rf.fit(X_train, y_train)

# Prediction
rf_prediction =  rf.predict(X_test)



# Calculate accuracy
accuracy = accuracy_score(y_test, rf_prediction)

# Calculate mean squared error (MSE)
mse = mean_squared_error(y_test, rf_prediction)

# Calculate root mean squared error (RMSE)
rmse = np.sqrt(mse)

# Print the results
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("MSE: {:.4f}".format(mse))
print("RMSE: {:.4f}".format(rmse))

```

```python
print(classification_report(y_test, lr_prediction, target_names= ['Not Readmitted', 'Readmitted']))
```

### Confusion Matrix

```python
fig, ax = plt.subplots(figsize=(12, 4))
plot_confusion_matrix(rf, X_test, y_test,
                      display_labels = ['Not Readmitted', 'Readmitted'],
                      cmap = 'Blues',
                      ax=ax
                    )
plt.title('Confusion Matrix of Random Forest Model \n', size = 15)
plt.show()
```
